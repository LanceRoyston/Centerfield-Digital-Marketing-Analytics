# Necessary Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
marketing_data = pd.read_csv("marketing.csv")

pd.option_context('mode.use_inf_as_na', True)

marketing_data.head()
```

# Exploratory Data Analysis (EDA)

```python
# Summary statistics for numerical data
numerical_summary = marketing_data.describe()
print("Summary Statistics for Numerical Data:")
print(numerical_summary)

# Summary statistics for categorical data
categorical_columns = marketing_data.select_dtypes(include=['object']).columns

print("Summary Statistics for Categorical Data:")
for col in categorical_columns:
    print(f"\nColumn: {col}")
    print(f"Unique Values: {marketing_data[col].nunique()}")
    print(f"Mode: {marketing_data[col].mode()[0]}")
    print("Value Counts:")
    print(marketing_data[col].value_counts())
```

# Distribution Visualizations

```python
# Creating Calculated Fields
marketing_data['CPC'] = marketing_data['mark_spent'] / marketing_data['clicks']
marketing_data['CTR'] = marketing_data['clicks'] / marketing_data['impressions']

# Distribution of Impressions
plt.figure(figsize=(10, 6))
plt.hist(marketing_data['impressions'].dropna(), bins=50, edgecolor='k', alpha=0.7)
plt.title('Distribution of Impressions')
plt.xlabel('Impressions')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# Distribution of Clicks
plt.figure(figsize=(10, 6))
plt.hist(marketing_data['clicks'].dropna(), bins=50, edgecolor='k', alpha=0.7)
plt.title('Distribution of Clicks')
plt.xlabel('Clicks')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# Distribution of Revenue
plt.figure(figsize=(10, 6))
plt.hist(marketing_data['revenue'].dropna(), bins=50, edgecolor='k', alpha=0.7)
plt.title('Distribution of Revenue')
plt.xlabel('Revenue')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
